```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib import rc 
rc('font', family='AppleGothic') 
plt.rcParams['axes.unicode_minus'] = False   
```

```python
df1 = pd.read_excel('2020년 화물차 위반유형별 교통사고.xls')
df1
```

```python
df1 = df1.fillna(0)
df1
```

```python
df2 = df1[['화물(건)', '사망자수']]
df2
```

```python
df_1 = df1["위반유형"]
df_2 = df1["발생건수"]
df_3 = df1["대형사고"]
df_4 = df1["화물(건)"]
df_5 = df1["사망자수"]
df_6 = df1["치사율(%)"]
```

```python
plt.figure(figsize=(15,10))
plt.bar(df_1, df_2)
plt.title('발생건수')
plt.show()
```
