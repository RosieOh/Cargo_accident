```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc 
rc('font', family='AppleGothic') 
plt.rcParams['axes.unicode_minus'] = False   
```

```python
df1 = pd.read_excel('2018-2020 화물차 연도별 교통사고.xls')
df1
```

```python
df1 = df1.fillna(0)
df1
```

```python
df2 = df1[['화물(건)', '사망자수']]
df2
```

```python
df_1 = df1["연도"]
df_2 = df1["발생건수"]
df_3 = df1["대형사고"]
df_4 = df1["화물(건)"]
df_5 = df1["사망자수"]
df_6 = df1["치사율(%)"]
```

```python
plt.bar(df_1, df_2)
plt.title('발생건수')
plt.show()
```

```python
plt.bar(df_1, df_3)
plt.title('대형사고')
plt.show()
```

```python
plt.bar(df_1, df_4)
plt.title('화물(건)')
plt.show()
```

```python
plt.bar(df_1, df_5)
plt.title('사망자수')
plt.show()
```

```python
plt.bar(df_1, df_6)
plt.title('치사율(%)')
plt.show()
```

```python
plt.plot(df_1, df_2)
plt.title('발생건수')
plt.show()
```

```python
p_1 = sns.distplot(df_2)
p_1.set_title("화물차 사고 발생 건수")
```

```python
sns.distplot(df_3)
```

```python
p = sns.lineplot(data = df2)
p.set_title("화물차 사망자 수 비교")
```

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(20,20))

sns.distplot(df1['발생건수'], ax=ax[0,0])
sns.distplot(df1['대형사고'], ax=ax[0,1])
sns.distplot(df1['사망자수'], ax=ax[1,0])
sns.distplot(df1['치사율(%)'], ax=ax[1,1])
```

```python
plt.plot(df_1, df_4, 'o-', color='#C8C8FF', label='line01')
plt.plot(df_1, df_5, '^--g', linewidth=5, markeredgewidth=5, label='line02')

plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('사망자수/치사율(%)')

plt.show()
```
