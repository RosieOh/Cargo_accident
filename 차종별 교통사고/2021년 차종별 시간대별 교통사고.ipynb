```python
df1.drop(['합계'], axis=1, inplace = True)
df1
```

```python
df2 = df1.iloc[9:12,1:14]
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df4 = df3.rename(columns=df3.iloc[0])
df5 = df4.iloc[1:,]
df5
```

```python
df6 = df5.reset_index().rename(columns={'index': '시간대별'})
df6
```

```python
plt.figure(figsize=(14,7))
plt.bar(df6["시간대별"], df6["사고건수"], width=0.5)
plt.title('시간대별 사고건수')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.bar(df6["시간대별"], df6["사망자수"], width=0.5)
plt.title('시간대별 사망자수')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.bar(df6["시간대별"], df6["부상자수"], width=0.5)
plt.title('시간대별 부상자수')
plt.show()
```

```python
x = df6["시간대별"]
y = df6["사고건수"]
a = df6["사망자수"]
b = df6["부상자수"]
```

```python
plt.figure(figsize=(14,7))
plt.plot(x, y, 'cs--', label = '사고건수') 
plt.plot(x, a, 'rx:', label = '사망자수')
plt.plot(x, b, 'yo--', label = '부상자수')
plt.title("시간대별 사고건수, 사망자수, 부상자수 비교")
plt.legend() 
plt.show()
```
