```python
plt.figure(figsize=(14,7))
plt.bar(df7["도로형태"], df7["사고건수"], width=0.5)
plt.title('사고건수')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.bar(df7["도로형태"], df7["사망자수"], width=0.5)
plt.title('사망자수')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.bar(df7["도로형태"], df7["부상자수"], width=0.5)
plt.title('부상자수')
plt.show()
```

```python
x = df7["도로형태"]
y = df7["사고건수"]
a = df7["사망자수"]
b = df7["부상자수"]
```

```python
plt.figure(figsize=(14,7))
plt.plot(x, y, 'cs--', label = '사고건수') 
plt.plot(x, a, 'rx:', label = '사망자수')
plt.plot(x, b, 'yo--', label = '부상자수')
plt.title("도로형태 별 사고건수, 사망자수, 부상자수 비교")
plt.legend() 
plt.show()
```
