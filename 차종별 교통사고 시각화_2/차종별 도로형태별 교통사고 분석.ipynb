```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
import os
```

```python
print(os.listdir('./'))
```

```python
data = pd.read_excel('./2021년 차종별 도로형태별 교통사고.xls')
data
```

```python
df = data.rename(columns=data.iloc[1])
df
```

```python
df = df.iloc[2:44,:]
df
```

```python
df.rename(columns={'사고년도':'건 수'},inplace=True)
df
```

```python
df.drop(['합계'], axis=1, inplace = True)
df
```

```python
df_lorry = df.iloc[9:12,:]
df_lorry
```

```python
df_lorry2 = df.iloc[9:12,1:]
df_lorry2.drop(['철길건널목'], axis=1, inplace = True)
df_lorry2
```

```python
df_lorry3 = df_lorry2.transpose()
df_lorry3
```

```python
df_lorry4 = df_lorry3.rename(columns=df_lorry3.iloc[0])
df_lorry4
```

```python
df_lorry5 = df_lorry4.iloc[1:,:]
df_lorry5
```

```python
df = df_lorry5.reset_index().rename(columns={"index": "도로형태별"})
df
```

```python
df.reset_index().rename(columns={"index": "id"})
```

```python
import matplotlib

matplotlib.rcParams['font.family'] ='Malgun Gothic'

matplotlib.rcParams['axes.unicode_minus'] =False
```

```python
plt.figure(figsize=(10,6))
plt.title("도로형태별 사고건수")
plt.xticks(rotation=45)
sns.barplot(x="도로형태별",y="사고건수",data=df)
```

```python
plt.figure(figsize=(10,6))
plt.title("도로형태별 사망자수")
plt.xticks(rotation=45)
sns.barplot(x="도로형태별",y="사망자수",data=df)
```

```python
plt.figure(figsize=(10,6))
plt.title("도로형태별 사고건수")
plt.xticks(rotation=45)
sns.barplot(x="도로형태별",y="부상자수",data=df)
```

```python
plt.figure(figsize=(10,6))
plt.title("사고건수 별 사망자수")
sns.regplot(x=df["사고건수"].astype(float),y=df["사망자수"].astype(float))
```

```python
plt.figure(figsize=(10,6))
plt.title("사고건수 별 부상자수")
sns.regplot(x=df["사고건수"].astype(float),y=df["부상자수"].astype(float))
```

```python
plt.figure(figsize=(10,6))
plt.title("도로형태별 건 수")
plt.xticks(rotation=45)
plt.plot(df['도로형태별'], df['사고건수'],"D:", label="사고건수") 
plt.plot(df['도로형태별'], df['사망자수'],"o--", label="사망자수")
plt.plot(df['도로형태별'], df['부상자수'],"*--", label="부상자수") 
plt.legend()
```
